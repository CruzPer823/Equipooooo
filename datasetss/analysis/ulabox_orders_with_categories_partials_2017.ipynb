## Inciso 5

Gráficas

```python
import seaborn as sns
sns.set_theme(style="white")
sns.heatmap(df.corr())
df.corr()
```

En esta mapa de calor podemos ver la relación que tiene cada categoria de la tabla, entre más claro el color quiere decir que esas categorias tienen mayor relación, por lo que podemos concluir con este mapa y un analisis, que tienen algo de relación los descuentos con las secciones de productos, pero no tienen mucha relación.

```python
g=sns.histplot(data=df, x="hour", multiple="stack", hue="weekday")
for q in df.hour.quantile([.25,.5,.75]):
    g.axvline(q, linestyle=":")
    g.text(q,2500,q)
```

En esta gráfica podemos observar la frecuencia con la que los clientes estuvieron comprando de acuerdo a la hora y día de la semana

```python
sns.displot(x="Food%", data=df)
sns.displot(x="Fresh%", data=df)
sns.displot(x="Drinks%", data=df)
sns.displot(x="Home%", data=df)
sns.displot(x="Beauty%", data=df)
sns.displot(x="Health%", data=df)
sns.displot(x="Baby%", data=df)
sns.displot(x="Pets%", data=df)
```

Como se puede apreciar en esta serie de histogramas, se puede ver el movimiento/venta que tienen algunos productos en cierta categoria. 

```python
sns.scatterplot(data=df, x="hour", y="total_items")
```
